```python
import matplotlib.pyplot as plt
```

```python
flu_case = [13, 14, 9, 16, 10, 18, 22, 19, 16, 22,
             24, 48, 34, 25, 17, 29, 33, 35, 28, 43,
             59, 44, 55, 63, 61, 48, 68, 61, 70, 76,
             78, 74, 87, 101, 120, 128, 105, 109, 120, 124,
             111, 128, 120, 133, 134, 139, 127, 130, 141, 147,
             439, 236, 218, 209, 213, 244, 329, 197, 351, 325]

days = range(1, len(flu_case)+1)
days_7d = range(7, len(flu_case)+1)  # from day_7 to day_end
```

```python
# 7 days moving average
# # first 7-day window, day 7 = average of day 1~ 7
# sum(flu_case[0:7])/7
flu_case_7d_avg = []
for day in days_7d:
    average_7d = sum(flu_case[day-7: day])/7
    flu_case_7d_avg.append(average_7d)
```

```python
plt.plot(days, flu_case, color='C0', linestyle="", marker="o", alpha=0.5, label = 'daily')
plt.plot(days_7d, flu_case_7d_avg, color='C1', linestyle="-", label = '7-day average')
plt.legend()
plt.xlabel('day')
plt.ylabel('covid cases')
```
